# Pregunta 3

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 10
p = 0.5

k_vls = np.arange(0, n+1)

probabilidades = stats.binom.pmf(k_vls, n , p)

plt.figure(figsize=(10,6))
plt.bar(k_vls, probabilidades, color="blue", alpha=0.7)

plt.xlabel("Nmero de caras")
plt.ylabel("Probabilidad")
plt.title("Distribucin binomial: 10 lanzamientos de una moneda sin trucos")
plt.xticks(k_vls)

plt.show()
```

# Pregunta 5

```python
media = 0
desviacion_estandar = 1

datosrndm = np.random.normal(loc=media, scale=desviacion_estandar, size=1000)

media_datos = np.mean(datosrndm)
desviacion_estandar = np.std(datosrndm)

datosrndm
```

# Pregunta 6

```python
media_datos = np.mean(datosrndm)

desviacion_datos = np.std(datosrndm)

mediana_datos = np.median(datosrndm)

print(f"La media de los datos generados es {media_datos}")
print(f"La desviacin estndar de los datos generadoos es {desviacion_datos}")
print(f"La mediana de los datos generados es de {mediana_datos}")
```

# Pregunta 7

```python
plt.figure(figsize=(10, 6))
plt.hist(datosrndm, bins=30, color="green", edgecolor="black", alpha=0.7, density=True)

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, media, desviacion_estandar)
plt.plot(x, p, "k", linewidth=2, label="Distribucin normal")

plt.xlabel("Valor")
plt.ylabel("Densidad de probabilidad")
plt.title("Histograma de datos generados y curva de distribucin normal")
plt.legend()

plt.show()
```

# Pregunta 8

```python
probabilidad_en_1 = stats.norm.cdf(1, loc=media, scale=desviacion_estandar)
probabilidad_menos_1 = stats.norm.cdf(-1, loc=media, scale=desviacion_estandar)

#Para hallar la probabilidad del rango solo lo restamos

probabilidad_rango = probabilidad_en_1 - probabilidad_menos_1

probabilidad_rango
```

```python
print(f"La probabilidad de que un valor est en el rango [-1,1] es de {probabilidad_rango}")
```
